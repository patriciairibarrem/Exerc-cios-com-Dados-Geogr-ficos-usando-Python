```python
# Importar bibliotecas GeoPandas, Pandas, Matplotlib pyplot, contextily 

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
#import contextily as ctx
```

```python
# Ler o arquivo neighbourhood
bairros = gpd.read_file(r'C:\Users\iribarre\Downloads\4+-+Data+Vizualization(1)\AssignmentData\neighbourhod_crimes.geojson')
bairros.head()
```

```python
# Ler o arquivo swimming pool
piscina = gpd.read_file(r'C:\Users\iribarre\Downloads\4+-+Data+Vizualization(1)\AssignmentData\swimming_pools.shp')
piscina.head()
```

```python
# Defina o novo CRS desejado como WGS 84
novo_crs = 'EPSG:4326'

# Converta o GeoDataFrame para o novo CRS
bairros = bairros.to_crs(novo_crs)
bairros.crs
```

```python
piscina.crs
```

```python
# Overlay o mapas
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
bairros.plot(ax=ax, color='#C05A3C', linewidth=0.3, alpha=0.5);
piscina.plot(ax=ax, 
                color='#F1ECEB',
                alpha=0.7)

# Adicione ttulo, cor, etc.
plt.title('Piscinas e Bairros', fontsize=40, fontname='Palatino Linotype', color='grey')
ax.axis('off')
plt.axis('equal')
plt.show()
```

```python
# Mude as cores para melhor visualizar o dado
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
bairros.plot(ax=ax, color='Grey', linewidth=0.3, alpha=0.5);
piscina.plot(ax=ax, 
                color='Black',
                alpha=0.7)

# Adicione ttulo, cor, etc.
plt.title('Piscinas nos Bairros', fontsize=40, fontname='Palatino Linotype', color='grey')
ax.axis('off')
plt.axis('equal')
plt.show()
```

```python
# Converta bairros em pontos de crimes atravs do .centroid
crimes = bairros.copy()

# Certifique-se de que os dados estejam no sistema de coordenadas certo (por exemplo, usando .to_crs())
bairros = bairros.to_crs('EPSG:4326')  # Se o CRS original for diferente, substitua 'EPSG:4326' pelo CRS correto

# Mostre as primeiras linhas do DataFrame criems
crimes.head()
```

```python
# Visualize uma coluna de atributo especfica
crimes.plot(column='Counts', figsize=(12, 10))
```

```python
# Normalize a coluna Count com a Shape_Area. Divida Counts por Shape_Area e multiplique por 100. 
crimes['crimes_perc'] = (crimes['Counts'] / crimes['Shape_Area'])*100
crimes['crimes_perc']
```

```python
# Use a nova coluna para visualizar a densidade de crimes
fig, ax = plt.subplots(figsize=(12,10))
crimes.plot(column='crimes_perc', cmap='Reds', scheme='quantiles', k=6, ax=ax, legend=True)
```

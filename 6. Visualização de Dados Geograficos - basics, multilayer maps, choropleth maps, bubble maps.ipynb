# 4. CHOROPLETH MAP (mapa coroplético)

Usa as 'cores' para distribuir quantitativamente os valores em áreas geográficas. 

* Desvantagem: as áreas não são uniformes e isso pode ocasionar que áreas grandes dominem a visualização.
   * Vantagem: Ótimo para representar densidades e proporção de quantidades usando cores.

```python
!pip install mapclassify
import mapclassify
```

```python
# Confira as colunas do GeoDataFrame
small_areas.columns
```

```python
# Visualize uma coluna de atributo específica
small_areas.plot(column='Age80_w', figsize=(12, 10))
```

GeoPandas usa MapClassify para classificar dados, então é preciso ter a biblioteca instalada.

```python
# Classifique os dados e as cores que os representam, e qualifique o schema - Ler a documentação
fig, ax = plt.subplots(figsize=(12,10))
small_areas.plot(column='Age80_w', cmap='Blues', scheme='quantiles', k=6, ax=ax, legend=True)
```

```python
# Normalize a população, contabilizando o total e derivando em porcentagens
small_areas['Age80_perc'] = (small_areas['Age80_w'] / small_areas['Total'])*100
small_areas['Age80_perc']
```

```python
# Use a nova coluna para visualizar a densidade populacional igual ou maior que 80 anos
fig, ax = plt.subplots(figsize=(12,10))
small_areas.plot(column='Age80_perc', cmap='Blues', scheme='quantiles', k=6, ax=ax, legend=True)
```

```python
# Mostre os mapas lado a lado para visualizar as differenças entre o mapa original e o normalizado
fig, ax = plt.subplots(1, 2, figsize=(20,10))
small_areas.plot(column='Age80_perc', cmap='YlGn', scheme='quantiles', k=6, ax=ax[0], legend=True)
ax[0].set_title('Normalized Choropleth Map')
small_areas.plot(column='Age80_w', cmap='YlGn', scheme='quantiles', k=6, ax=ax[1], legend=True)
ax[0].set_title('Row Data Choropleth Map')
plt.show()
```

# 5. BUBBLE MAP (mapa de bolhas)
Usa o 'tamanho' como codificação de variáveis. O tamanho do circulo representa o valor númerico numa área geográficas.

```python
# Converta small_areas em pontos através do .centroid
small_areas_points = small_areas.copy()
small_areas_points['geometry'] = small_areas_points['geometry'].centroid
small_areas_points.head()
```

```python
small_areas_points.plot(color='grey')
```

```python
small_areas_points.plot(color='grey', markersize=5)
```

```python
fig, ax = plt.subplots(figsize=(16, 16))
small_areas_points.plot(ax=ax, color='grey', markersize='Age80_w')

#ax.axis('off')
plt.axis('equal')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 16))
small_areas.plot(ax=ax, color='lightgray', edgecolor='grey', linewidth=0.4)
small_areas_points.plot(ax=ax, color='#07424A', markersize='Age80_w', alpha=0.7, categorical=False, legend=True)

ax.axis('off')
plt.axis('equal')
plt.show()
```
